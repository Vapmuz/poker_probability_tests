PROBABILITY OF HAVING A COUPLE WITH TWO DIFFERENT VALUES IN THE HAND MADE WITH X DECKS:


```python
def is_couple( number_one, number_two, card_in_pool,nr_decks=2): #nr_decks,number_one,number_two, card_in_pool
    """I'm checking if the numbers 3 and 2 are in the next 5 cards, if  one condition is True then it returns 1"""
    my_deck = cd.Deck(nr_decks)
    hand = (str(number_one), str(number_two))

    my_deck.shuffle()
    for x in range(card_in_pool):
        my_deck.deal_card()

    for cards in my_deck.pool:
        for x in hand:
            if cards.rank == x:
                return 1
    return(0)

def check_couple(number_one, number_two, card_in_pool):
    good_tries = 0
    tries = 0
    for x in range(10000):
        good_tries = good_tries + is_couple(number_one, number_two, card_in_pool)
        tries = tries + 1
    print(good_tries/tries)
    
```

```python
check_couple(2,3,5)
```

We can clearly see that with a pool of 10'000 tries the probability converges to 0.57...
so I can clearly say that the probabilty of having at least a couple is 0.57.

```python
"""functions"""
def couple_at_beginning():
    my_deck = cd.Deck(2)
    my_deck.shuffle()
    for x in range(2):
        my_deck.deal_card()
    if my_deck.pool[0].rank == my_deck.pool[1].rank:
        return(1)
    else:
        return(0)
    
    
def is_couple_at_beginning():
    a=0
    b=0
    for x in range(10000):
        a= a+couple_at_beginning()
        b= b+1
    print(a/b)
```

Then what is the absolute possibility of having a couple?



```python
"""functions"""
def couple_absolute(nr_card_dealt):
    my_deck = cd.Deck(2)
    my_deck.shuffle()
    for x in range(nr_card_dealt):
        my_deck.deal_card()
    a = -1
    for x in my_deck.pool:
        my_list = my_deck.pool.copy()
        my_list.remove(x)
        for y in range(len(my_list)):
            if x.rank == my_list[y].rank :
                return(1)
    return(0)

def couple_absolute_possibility(nr_card_dealt):
    a=0
    b=0
    for x in range(10000):
        a= a+couple_absolute(nr_card_dealt)
        b= b+1
    print(str(a)+" over " +str(b))
    print(a/b)
    

```
